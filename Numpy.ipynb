## **np.nditer use in loop for numpy array**

```python
import numpy as np
arr = np.array([10, 20, 30, 40, 53, 79, 50, 60, 70])
for i in np.nditer(arr):
  if i%2==0:
    print(i,end=" ")
```

**np.ndenumerate use to print elements with their index using loop in numpy array**

```python
import numpy as np
arr = np.array([[10, 20, 30, 40],
                 [53, 79, 50, 60]])
for ind, i in np.ndenumerate(arr):
  print(ind, i)
```

# **View & Copy**
**View**: The view of original array, if we change the view the original will also changed.

**Copy**: It will create the copy so that orginal will remains the same. For that we use np.copy().

```python
import numpy as np
arr = np.array([1,2,3,4,5])
new = arr[0:3].copy()
print(arr)
print(f"New: {new}")
new[0] = 100
print(arr)
print(f"New: {new}")
```

# Concatination
Combining two arrays using np.concatenate() function

**np.vstack([arr1, arr2])**

**np.hstack([arr1, arr2])**


```python
a = np.array([1,2,3])
b = np.array([10,20,40])
c = np.concatenate([a,b])
print(c)
```

```python
# hstack --> use to combine two arrays horrizontaly
import numpy as np
a = np.array([[1,2,3], [4,5,6]])
b = np.array([[10,20,40],[50, 60, 70]])
print(a)
print(b)
print("Horizontally: ")
print(np.hstack([a, b]))

print("Vertically: ")
print(np.vstack([a, b]))
```

```python
import numpy as np
b = np.array([[10,20,40],[50, 60, 70]])
a = np.array([[1,2,3], [4,5,6]])
print(np.stack((a, b), axis=0)) # axis = 0  means vertically
print("Lolaa")
print(np.stack((a, b), axis=1))# axis = 0  means horizontally
```

```python
import numpy as np
sales = np.array([
    [1200, 1100, 1000, 900],
    [1500, 1450, 1400, 1300],
    [1600, 1580, 1550, 1500],
    [1700, 1650, 1620, 1600],
    [1800, 1750, 1700, 1680],
    [2000, 1950, 1900, 1850]
])
print("Total sales of prodeuct 1 ")
print(np.sum(sales[:,0]))
best = np.sum(sales, axis=1)
print(best)
print("Best Month")
print(f"Sales: {np.max(best)} Month: {np.argmax(best)+1}")
print("Average of all 6 months")
average = np.mean(sales)
print(average)
```

```python
data = np.array([5, 7, 3, 100, 6, 7, 120, 8])
new_data = data[data<50]
print("After Removing outliers")
print(new_data)
mean = np.mean(new_data)
print(f"Mean: {mean}")
data[data>50]=mean
print(data)
```
